```python
def low_pbr(cluster,period,measure): #Cluster, 투자 시작일, 투자 기간, PBR 기준
    jongmok=list(data[(data['클러스터']==cluster) & (data['일자']=='2016년 2월')& (data['PBR']<measure)]['종목명'])
    up_down=[]
    for i in range(len(jongmok)):
        price=[]
        for j in range(0,len(close),period):
            price.append(close[jongmok[i]][j])
        for k in range(1,len(price)):
            up_down.append((price[k]-price[0])/price[0])
    up_down=pd.DataFrame(up_down)    
    return up_down.median()[0]
```

```python
low_pbr(2,240,10)
```

```python
#grid search
low_pbr_cluster=[]
low_pbr_measure=[]
low_pbr_period=[]
low_pbr_ary=[]
for i in range(3):#cluster i 
    for j in range(10,1150,30):#period j
        for k in range(5,20): #measure k
            low_pbr_cluster.append(i)
            low_pbr_measure.append(0.1*k)
            low_pbr_period.append(j)
            low_pbr_ary.append(low_pbr(i,j,0.1*k))
            

low_pbr_df=pd.DataFrame({'클러스터':low_pbr_cluster,'PBR_기준':low_pbr_measure,'투자기간':low_pbr_period,'결과':low_pbr_ary})
low_pbr_df
```

## 3.과매도로 인한 대량 낙폭 발생 시 매수 전략

```python
def oversold(cluster,k): # Cluster 에 대해 k 일간 하락 시 매수 전략, 경우의 수 총 1160*719 = 834040개
    jongmok=list(data[(data['클러스터']==cluster)& (data['일자']=='2016년 2월')]['종목명'])
    up_down=[]
    for i in range(len(jongmok)):
        for j in range(1,len(close)):
            cnt=0
            for days in range(k+1):
                try:
                    if close[jongmok[i]][j+days]<close[jongmok[i]][j+days-1]:
                        cnt+=1
                except:
                    pass
            if cnt==k:
                try:
                    up_down.append((close[jongmok[i]][j+k]-close[jongmok[i]][j+k-1])/close[jongmok[i]][j+k-1])
                    j+=k
                except:
                    pass
    up_down=pd.DataFrame(up_down) 
    return up_down.median()[0]
    #return up_down
    
```

```python
#grid search
oversold_cluster=[]
oversold_period=[]
oversold_ary=[]
for i in range(3):#cluster i 
    for j in range(1,10):#period j
        oversold_cluster.append(i)
        oversold_period.append(j)
        oversold_ary.append(oversold(i,j))
            

oversold_df=pd.DataFrame({'클러스터':oversold_cluster,'하락기간':oversold_period,'결과':oversold_ary})
oversold_df
```

```python
oversold_df.to_csv('oversold.csv',encoding='ANSI')
```

## 4.볼린저밴드 폭 축소, 하한선 터치 후 상승 시 매수

```python
def bb(cluster, w=20, k=2,period=1): # w-day bollinger band with k-sigma standard deviation
    #period 일 간 횡보(bb 폭 축소) 후 하한선 터치, 이후 상승 시 매수
    jongmok=list(data[(data['클러스터']==cluster)& (data['일자']=='2016년 2월')]['종목명'])
    up_down=[]
    for i in range(len(jongmok)):
        x=pd.Series(close[jongmok[i]])
        mbb = x.rolling(w).mean()
        ubb = mbb + k * x.rolling(w).std()
        lbb = mbb - k * x.rolling(w).std()
        x=x[w:]
        mbb=mbb[w:]
        ubb=ubb[w:]
        lbb=lbb[w:]
        for j in range(1,len(mbb)-1):
            if x[j]>lbb[j] and x[j-1]<lbb[j-1]:
                up_down.append((x[j+1]-x[j])/x[j])
    up_down=pd.DataFrame(up_down)
    return up_down.median()[0]
```

```python
#grid search
bb_cluster=[]
bb_w=[]
bb_k=[]
bb_ary=[]
for i in range(3):#cluster i 
    for j in range(1,10):#period j
        bb_cluster.append(i)
        bb_period.append(j)
        bb_ary.append(bb(i,j))
            

bb_df=pd.DataFrame({'클러스터':bb_cluster,'하락기간':bb_period,'결과':bb_ary})
bb_df
```

```python
def ma(cluster, w=20): # w-day moving average 
    jongmok=list(data[(data['클러스터']==cluster)& (data['일자']=='2016년 2월')]['종목명'])
    up_down=[]
    for i in range(len(jongmok)):
        x=pd.Series(close[jongmok[i]])
        ma = x.rolling(w).mean()
        x=x[w:]
        ma=ma[w:]
        for j in range(1,len(ma)-1):
            if x[j]>ma[j] and x[j-1]<ma[j-1]:
                up_down.append((x[j+1]-x[j])/x[j])
    up_down=pd.DataFrame(up_down)
    print(up_down.describe())
```
